#1.JOIN

##1.1 INNER JOIN
INNER JOIN(内部結合)はそれぞれのテーブルのしていしたカラム(列)の値が一致するものだけを結合します

```
SELECT
 (*)
FROM
 (Table1)
INNER JOIN
 (Table2)
ON
 (結合の条件)
```

###例：

````python
%%sql
SELECT
 *
FROM
 film
````

````python
%%sql
SELECT
 *
FROM
 store
````

````python
%%sql
SELECT
 *
FROM
 store
INNER JOIN
 customer
ON
 store.store_id=customer.store_id
````

##1.2 OUTER JOIN

外部結合は、内部結合のようにそれぞれのテーブルの指定したカラム値(列の値)が一致するものを結合するのに加えて、どちらかのテーブルにしか存在しないものに関しても取得します。LEFT  OUTER JOINは左側のテーブルを軸にして外部結合を行い、RIGHT OUTER JOINは右側のテーブルを軸にして外部結合を行います。


###1.2.1 LEFT OUTER JOIN

```
SELECT
 (*)
FROM
 (Table1)
LEFT OUTER JOIN
 (Table2)
ON
 (結合の条件)
```

####例：

````python
%%sql
SELECT
 fc.category_id, fc.last_update, f.rental_rate, f.rating
FROM
 film_category fc
LEFT OUTER JOIN
 film f
ON
 fc.film_id=f.film_id
ORDER BY
category_id
````

###1.2.2 RIGHT OUTER JOIN

```
SELECT
 (*)
FROM
 (Table1)
RIGHT OUTER JOIN
 (Table2)
ON
 (結合の条件)
```

````python
%%sql
SELECT
 fc.category_id, fc.last_update, f.rental_rate, f.rating
FROM
 film_category fc
RIGHT OUTER JOIN
 film f
ON
 fc.film_id=f.film_id
ORDER BY
category_id
````

OUTER JOINを使用することで一致しないデータも含めて関連性のあるデータを取得できます。
データの補完や欠損データの確認に役立ちます。

#2. VIEW

テーブルの特定部分や複数テーブルを結合し、ユーザに必要な部分だけをあたかも一つのテーブルであるかのように見せるための、仮想的なテーブルの概念です。


CREATE VIEW (ビュー名) AS (定義)

##例:

````python
%%sql
SELECT
 *
FROM
 customer
````

````python
%%sql
CREATE VIEW customer_email AS
SELECT
first_name, last_name, email
FROM
customer
````

````python
%%sql
SELECT
*
FROM
customer_name
````

テーブルの内容を変更したい場合、

**CREATE OR REPLACE VIEW ビュー名 AS 定義**

を用いて変更します。

##例:

````python
%%sql
CREATE OR REPLACE VIEW customer_name AS
SELECT
first_name, last_name, email
FROM
customer;
````

````python
%%sql
SELECT
*
FROM
customer_name
````

作ったテーブルを消去するためには、

DROP VIEW テーブル名

で消去します。

##例:


````python
%%sql
DROP VIEW customer_name
````

````python
%%sql
SELECT
*
FROM
customer_name
````
