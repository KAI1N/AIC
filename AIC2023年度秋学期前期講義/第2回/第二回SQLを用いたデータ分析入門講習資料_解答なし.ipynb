## 1. ORDER BY

- ORDER 句

**ORDER BY**
- テーブルからSELECTでデータを取得するとき,「ORDER BY」を使うと、指定されたカラムを基準に並べ替えることができます。
- 複数指定することもでき、前に書いたカラムを優先してソートします
- 昇順（ASC）または降順（DESC）二つのソート方法があります。
- Nan, nullは一番大きな値として扱われます

````python
%%sql
-- population で ORDER BY してみよう
SELECT
  *
FROM
  countries
ORDER BY
  population
````

````python
%%sql
-- population で 昇順に ORDER BY してみよう
SELECT
  *
FROM
  countries
ORDER BY
  population ASC
````

````python
%%sql
-- population で 降順に ORDER BY してみよう
SELECT
  *
FROM
  countries
ORDER BY
  population DESC
````

````python
%%sql
-- population で 複数のカラムで ORDER BY してみよう
SELECT
  *
FROM
  countries
ORDER BY
  continent, -- 名前等の場合はアルファベット順になる
  population ASC
````

## 2. GROUP BY / HAVING / 集計関数

- GROUP 句

**GROUP BY**
  - 指定したカラムでグルーピングを行う
  - SELECTの中には<b>指定したカラム</b>と<b>集計関数</b>しか使えない

- 集計関数
  - 集計をするための関数

|集計関数|動作|
|---|---|
|COUNT(X)|カラムXの件数を返す、```*```で指定するとレコード数を返す|
|SUM(X)|カラムXの合計値を返す|
|AVG(X)|カラムXの平均値を返す|
|MAX(X)|カラムXの最大値を返す|
|MIN(X)|カラムXの最小値を返す|

- HAVING 句

**HAVING BY**
- `GROUP BY`句でグループ化した結果に対して条件を適用するための句。
- `WHERE`句は集約前の各行に対して条件を適用するのに対し、`HAVING`句は集約後のグループ全体に対して条件を適用するもの。
- 例として特定の条件を満たすグループだけを結果として取得する場合に使用するとかがあります。



### 1.1 COUNT

````python
%%sql
-- ヨーロッパの国数(正確にはEuropeとフラグがあるカラム数)
SELECT
  COUNT(*) as cnt
FROM
  countries
WHERE
  continent = 'Europe'
````

````python
%%sql
-- 各大陸の国数 エラーになる例

SELECT
  continent,
  COUNT(*) as cnt
FROM
  countries
````

````python
%%sql
-- 各大陸の国数 正常実行
SELECT
  continent,
  COUNT(*) as cnt
FROM
  countries
GROUP BY
  continent
````

### DISTINCT

**DISTINCT**
- 重複するデータを除去して、一意のデータだけを取得するためのキーワード。
- 一つ以上のカラムに適用することで、特定のカラムの組み合わせが一意になるように結果を取得することもできる。

````python
%%sql

SELECT
  COUNT(DISTINCT continent) continent_cnt,
  COUNT(continent) as cnt, -- 名前の付けたいときは as　アリ/ナシどちらでも良い
FROM
  countries
````

### 1.2 平均 AVG(column) ・ 1.3 合計 SUM(column)

````python
%%sql
-- ヨーロッパ内での平均,合計 人口/GDP
SELECT
  AVG(population) as avg_population,
  AVG(gdp) as avg_gdp,
  SUM(population) as sum_population,
  SUM(gdp) as sum_gdp
FROM
  countries
WHERE
  continent = 'Europe'
````

### 1.4 MAX, MIN 関数

````python
%%sql
-- GDP最高、最低値
SELECT
  Max(gdp),
  Min(gdp)
FROM
  countries
````

### 1.4 Having

````python
%%sql
-- 各大陸の国数(10カ国以上) 正常実行
SELECT
  continent,
  COUNT(*) AS country_count
FROM
  countries
GROUP BY
  continent
HAVING
  COUNT(*) > 10
````

## 3. CASE
- <b>CASE</b>
  - SELECTの中で使う
  - 条件に応じた新しいカラムを宣言することができる
  - イメージとしてはIF文に近い

```sql
SELECT
  CASE
    WHEN (条件式1) THEN (出力1)
    WHEN (条件式2) THEN (出力2)
    WHEN (条件式3) THEN (出力3)
  END AS (カラム名)
FROM (テーブル名)
```

````python
%%sql

SELECT
  name,
  area,
  CASE
      WHEN area < 100000 THEN 'Small'
      WHEN area BETWEEN 100000 AND 500000 THEN 'Medium'
      ELSE 'Large'
  END as size_category
FROM countries;
````

## 3. Window 関数

### ウィンドウ関数と集計関数の違い
- ウィンドウ関数には、AVG・SUM・MAX・MINなどの集計関数と同じ関数がある。
- 集計関数との違いは、出力結果が異なる。
- 集計関数は、グループごとの、レコードの集約を行なった１行の結果を返す。
- ウィンドウ関数では、レコードの集約は行わず、入力と同じ数のレコードを返す。


### 3.1 Window関数で用いる句

#### 3.1.1 OVER句

**OVER**
- ウィンドウ関数を適用する範囲を指定する。
- PARTITION BYやORDER BYと組み合わせて使用することが多い。


---

#### 3.1.2 PARTITION BY句

**PARTITION BY**
- OVER句内で使用し、データを特定のカラムの値に基づいてグループ化（パーティション化）する。
- 各パーティション内でウィンドウ関数を独立して適用する。


---

#### 3.1.3 ORDER BY句

**ORDER BY**
- OVER句内で使用し、ウィンドウ内の行の順序を指定する。
- 例えば、LAG/LEADやRANKなどの関数で順序が重要な場合に使用する。

```sql
SELECT
  column_name,
  SUM(column_name) OVER () AS sum
  SUM(column_name) OVER (PARTITION BY partition_column)  AS partition_sum
  SUM(column_name) OVER (ORDER BY order_column)  AS order_sum
  SUM(column_name) OVER (PARTITION BY partition_column ORDER BY order_column) AS order_partition_sum
FROM table_name;
```


````python
%%sql
-- 大陸ごとのGDP最高、最低値　平均値をその国の名前、大陸名等と一緒に一括取得
SELECT
  name,
  continent,
  gdp,
  AVG(gdp) OVER (PARTITION BY continent),
  Max(gdp) OVER (PARTITION BY continent),
  Min(gdp) OVER (PARTITION BY continent)
FROM
  countries
GROUP BY
  continent, name, gdp
````

### 3.2 集計関数を用いたWindow関数

### 3.3 集計関数で登場したもの以外の主なWindow関数
#### 3.3.1. LAG/LEAD

**LAG/LEAD**
- 現在の行の前後の行の値を取得する。
- LAGは前の行、LEADは次の行の値を返す。

```sql
SELECT
  LAG(column_name) OVER (ORDER BY order_column) AS prev_value,
  column_name,
  LEAD(column_name) OVER (ORDER BY order_column) AS next_value
FROM
  table_name;

```

````python
%%sql

SELECT
  name,
  LAG(name) OVER(ORDER BY id) as prev_stop,
  LEAD(name) OVER(ORDER BY id) as next_stop
FROM
  stops;
````

#### 3.3.2. RANK

**RANK**
- 各行に順位を付ける。
- 同じ値の行は同じ順位を持ち、次の順位はスキップされる。

```sql
SELECT
  column_name,
  RANK() OVER (ORDER BY order_column DESC) AS rank
FROM
  table_name;
```

````python
%%sql
SELECT
  name,
  continent,
  gdp,
  RANK() OVER(PARTITION BY continent ORDER BY gdp DESC) as rank_within_continent
FROM
  countries
WHERE
  gdp IS NOT NULL
````

#### 3.3.3. ROW_NUMBER

**ROW_NUMBER**
- 各行に一意の順序を付ける。
- 同じ値でも順序はスキップされず、連続する番号が付けられる。

```sql
SELECT
  column_name,
  ROW_NUMBER() OVER (ORDER BY order_column DESC) AS row_num
FROM
  table_name;
```

````python
%%sql
SELECT
  title,
  rdate,
  ROW_NUMBER() OVER(ORDER BY rdate) as row_num
FROM
  albums
````

### 4. その他
SQLの文法は順番が決まっています．
1. <b>select</b>　⇒　列(カラム)を指定
1. <b>from</b>　⇒　テーブルを指定
1. join　⇒　複数テーブルの結合処理（第3回で扱います）
1. <b>where</b>　⇒　絞り込み条件の指定
1. <b>group by</b>　⇒　グループ条件の指定
1. <b>having</b>　⇒　グループ化後の絞り込み条件の指定（今回は扱いません、whereと似た処理）
1. <b>order by</b>　⇒　ソート順の指定
1. limit　⇒　取得する行数の指定
